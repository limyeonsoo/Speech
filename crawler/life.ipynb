```python
import json
import os
def kewyord_check(dictionary_check, keyword):
    with open("dictionary_check","r") as f:
        check = json.load(f)
    for key in check.keys():
        if keyword in key:
            return True
    return False
        
def file_check(dictionary_check, file):
    with open(dictionary_check,"r") as f:
        check = json.load(f)
    for wav in check.values():
        if file in wave:
            return True
    return False

def keyword_10(path):
    if os.listdir(path) == None:
        return False
    elif len(os.listdir(path)) >= 10:
        return True
    else:
        return False
```

```python
import os
from tqdm import tqdm
import subprocess
import glob
import json

save = "/mnt/sdb/YS/youtube_data/일상생활/"
cmd = "ffmpeg -i "

count = 0
dictionary_index = {}
dictionary_check = {}
# with open("dictionary_check","r") as f:
#     dictionary_check = json.load(f)
# with open("dictionary_index", "r") as f:
#     dictionary_index = json.load(f)
for root, directory, files in os.walk('/home/knlab/Summer_study/Lab_members/JYC/honk/keyword_spotting_data_generator/일상생활/'):
    for direc in tqdm(directory):
        count+=1
        if count%100 == 0: print(count)
        if keyword_10(root+direc): # 파일이 10개 이상.
            waves = [];check = [];count = 0;
            for file in tqdm(os.listdir(root+direc)):
                if file in dictionary_check.values():
                    continue
                else:  #없는 파일
                    if file.endswith('wav'):
                        file_path = root+direc+"/"+file
                        waves.append(file_path)
                        check.append(file)
                        if not os.path.isdir(save+direc):
                            os.makedirs(save+direc)
                        subprocess.call(cmd+file_path+' '+os.path.join(save,direc+"/"+file).replace("wav","flac"),shell=True)         
            dictionary_index[direc] = len(waves)
            dictionary_check[direc] = check
        else:
            continue
with open("dictionary_index_일상생활","w") as f:
    f.write(json.dumps(dictionary_index))
with open("dictionary_check_일상생활", "w") as f:
    f.write(json.dumps(dictionary_check))
    
```

```python
count = 0
save = "/mnt/sdb/YS/youtube_data/일상생활/"
dictionary_index = {}
dictionary_check = {}

for root, directory, files in os.walk('/home/knlab/Summer_study/Lab_members/JYC/honk/keyword_spotting_data_generator/일상생활/'):
    label = root.split('/')[-1]
    if len(files) >= 10:
        for file in tqdm(files):
            if file.endswith(".wav"):
                data, samplerate = sf.read(os.path.join(root+"/"+file))
                if not os.path.isdir(save+label):
                    os.makedirs(save+label)
                sf.write(save+label+"/c_"+file[:-3]+".wav", data, samplerate)
        dictionary_index[label] = len(files)
        dictionary_check[label] = files
with open("dictionary_index_일상생활","w") as f:
    f.write(json.dumps(dictionary_index))
with open("dictionary_check_일상생활", "w") as f:
    f.write(json.dumps(dictionary_check))                
            
```
